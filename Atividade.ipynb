```python
import os

base_path = "/content/processamento-imagens/Imagens"

# Lista todas as imagens da pasta
imagens = os.listdir(base_path)

print("Imagens encontradas:", imagens)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_path = os.path.join(base_path, "imagem1.jpg")

img = cv2.imread(img_path)

# Verificação
if img is None:
    print("Erro ao carregar a imagem. Verifique o nome/caminho.")
else:
    print("Imagem carregada com sucesso!")
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Imagem carregada do GitHub")
plt.axis('off')
```

```python
imgs = []

for nome in imagens:
    caminho = os.path.join(base_path, nome)
    imagem = cv2.imread(caminho)

    if imagem is not None:
        imgs.append(imagem)

print(f"{len(imgs)} imagens carregadas com sucesso!")
```

```python
for i, imagem in enumerate(imgs):
    plt.figure()
    plt.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB))
    plt.title(f"Imagem {i+1}")
    plt.axis('off')
```

```python
# EXERCÍCIO 2

print("Shape:", img.shape)
print("Size:", img.size)
print("Dtype:", img.dtype)
```

```python
# EXERCÍCIO 3

cv2.imwrite('imagem_convertida.jpg', img)
```

```python
# EXERCÍCIO 4

h, w, _ = img.shape
centro = img[h//2, w//2]
print("Pixel central (BGR):", centro)
```

```python
# EXERCÍCIO 5

img_mod = img.copy()
img_mod[0:100, 0:100] = (0, 255, 0)
```

```python
# EXERCÍCIO 6

roi = img[100:300, 100:300]

plt.imshow(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB))
plt.title("Recorte (ROI)")
plt.axis('off')
```

```python
# EXERCÍCIO 7

b, g, r = cv2.split(img)
```

```python
# EXERCÍCIO 8

plt.imshow(g, cmap='gray')
plt.title("Canal Verde")
plt.axis('off')
```

```python
# EXERCÍCIO 9

img_merge = cv2.merge([r, g, b])
plt.imshow(img_merge)
plt.title("Canais invertidos (R ↔ B)")
plt.axis('off')
```

```python
# EXERCÍCIO 10

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.title("Escala de Cinza")
plt.axis('off')
```

```python
# EXERCÍCIO 11

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
s = hsv[:, :, 1]

plt.imshow(s, cmap='gray')
plt.title("Canal Saturação (S)")
plt.axis('off')
```

```python
# EXERCÍCIO 12

bright = cv2.add(img, np.ones(img.shape, dtype='uint8') * 50)

plt.imshow(cv2.cvtColor(bright, cv2.COLOR_BGR2RGB))
plt.title("Imagem com brilho aumentado")
plt.axis('off')
```

```python
# EXERCÍCIO 13

img2 = cv2.resize(img, (img.shape[1], img.shape[0]))

blend = cv2.addWeighted(img, 0.5, img2, 0.5, 0)

plt.imshow(cv2.cvtColor(blend, cv2.COLOR_BGR2RGB))
plt.title("Mesclagem de imagens")
plt.axis('off')
```

```python
# EXERCÍCIO 14

neg = 255 - gray

plt.imshow(neg, cmap='gray')
plt.title("Negativo")
plt.axis('off')
```

```python
# EXERCÍCIO 15

mask = np.zeros(gray.shape, dtype='uint8')
cv2.circle(mask, (w//2, h//2), 100, 255, -1)

res = cv2.bitwise_and(img, img, mask=mask)

plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
plt.title("Máscara circular")
plt.axis('off')
```

```python
# EXERCÍCIO 16

small = cv2.resize(img, None, fx=0.3, fy=0.3)
print("Novo shape:", small.shape)
```

```python
# EXERCÍCIO 17

hist = cv2.calcHist([gray], [0], None, [256], [0, 256])
print("Histograma (primeiros valores):")
print(hist[:10])
```

```python
# EXERCÍCIO 18

colors = ('b', 'g', 'r')

for i, col in enumerate(colors):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist)

plt.title("Histograma RGB")
plt.show()
```

```python
# EXERCÍCIO 19

eq = cv2.equalizeHist(gray)

plt.imshow(eq, cmap='gray')
plt.title("Equalização de Histograma")
plt.axis('off')
```

```python
# EXERCÍCIO 20

_, th = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(th, cmap='gray')
plt.title("Threshold binário")
plt.axis('off')
```
